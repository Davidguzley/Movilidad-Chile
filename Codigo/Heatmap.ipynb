```python
import pandas as pd
import numpy as np

#df = pd.read_csv('matriz_viajes_v2.csv')
df = pd.read_csv('Matriz.csv')
```

```python
df
```

```python
df = df.drop(['Unnamed: 0'], axis = 1)
```

```python
df
```

```python
df['Concatenados'] = df['Comuna_Origen'] + ', ' + df['Comuna_Destino']
```

```python
df
```

```python
len(df['Concatenados'].unique())
```

```python
viajes = df['Concatenados'].value_counts()
```

```python
viajes[0]
```

```python
viajes
```

```python
df2 = viajes.to_frame()
```

```python
df2 = df2.reset_index()
```

```python
df2
```

```python
aux = {
    'Origen': [],
    'Destino': [],
    'Viajes': []
}

for i in range(len(df2)):
  x = df2['index'][i].split(", ")
  aux['Origen'].append(x[0])
  aux['Destino'].append(x[1])
  aux['Viajes'].append(df2['Concatenados'][i])
```

```python
len(aux['Destino'])
```

```python
len(aux['Origen'])
```

```python
len(aux['Viajes'])
```

```python
matriz = pd.DataFrame.from_dict(aux)
```

```python
matriz.to_csv('heatmap.csv')
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set()

matriz = matriz.pivot("Origen", "Destino", "Viajes")

sns.color_palette("flare", as_cmap=True)

ax = sns.heatmap(matriz, cmap="crest")
plt.title("Heatmap Viajes")
plt.show()
```
