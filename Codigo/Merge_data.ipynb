```python
import pandas as pd
import numpy as np
df_antenas = pd.read_csv('antenas.csv')
df = pd.read_csv('loreto_idea_clean.csv')
```

```python
df
```

```python
df["timestamp"] = pd.to_datetime(df["timestamp"])
```

```python
def tiemp(t1, t2):
  return (t1 - t2) / pd.Timedelta(hours=1)
```

```python
df['tiempo_min'] = (tiemp(df['timestamp'], df.groupby(['PHONE_ID'])['timestamp'].shift(1))) * 60
```

```python
df['lat'] = df['lat'].astype("string")
df['lon'] = df['lon'].astype("string")
df['coordinates'] = df['lat'] + ',' + df['lon']
```

```python
df_matriz = df[['PHONE_ID','timestamp', 'coordinates', 'tiempo_min',]]
```

```python
df_matriz
```

```python
df_merge = df_matriz.merge(df_antenas, left_on='coordinates', right_on='coordinates')
df_merge
```

```python
df_merge = df_merge.sort_values(by = ['PHONE_ID', 'timestamp'])
```

```python
df_merge = df_merge.reset_index()
df_merge
```

```python
df_merge = df_merge.drop(['index'], axis = 1)
```

```python
df_merge
```

```python
df_merge = df_merge.drop(['Unnamed: 0'], axis = 1)
```
