```python
import pandas as pd
import numpy as np

df = pd.read_csv('new_train.csv')
df_antenas = pd.read_csv('antenas.csv')
```

```python
df = df.drop(['Unnamed: 0', 'PHONE_ID'], axis = 1)
```

```python
df
```

```python
df['HomeOffice?'].value_counts()
```

```python
df_antenas = df_antenas.drop(['Unnamed: 0'], axis = 1)
```

```python
df_antenas
```

```python
df_merge = df.merge(df_antenas, left_on='Trabajo', right_on='coordinates')
df_merge
```

```python
df_merge = df_merge.drop(['Trabajo', 'Casa', 'coordinates'], axis = 1)
```

```python
df_merge
```

```python
df_merge['HomeOffice?'].value_counts()
```

```python
dict_aux = {}

for i in range(len(df_merge)):
    if df_merge['comuna'][i] in dict_aux.keys():
        if df_merge['HomeOffice?'][i] == 0:
            dict_aux[df_merge['comuna'][i]][0] += 1
        else: 
            dict_aux[df_merge['comuna'][i]][1] += 1
    else:
        if df_merge['HomeOffice?'][i] == 0:
            dict_aux[df_merge['comuna'][i]] = [1, 0]
        else: 
            dict_aux[df_merge['comuna'][i]] = [0, 1]
```

```python
dict_aux
```

```python
for i in dict_aux.values():
    i.append(i[1] * 100 / (i[0] + i[1]))
```

```python
dict_aux
```

```python
df_home = pd.DataFrame.from_dict(dict_aux, orient='index')
```

```python
df_home
```

```python
df_home = df_home.drop([0, 1], axis = 1)
```

```python
df_home
```

```python
df_home = df_home.rename(columns={2: '%_of_homeoffice'})
```

```python
df_home
```

```python
df_home.to_csv('HomeOffice.csv')
```
