### __Merge data to create our model__

```python
import numpy as np
import pandas as pd
```

```python
df0 = pd.read_csv("datasets_merge/HomeOffice.csv")
df0.head()
```

```python
df0["Unnamed: 0"] = df0["Unnamed: 0"].str.replace(" de ", " De ")
df0
```

```python
df1 = pd.read_csv("datasets_merge/conexiones_fijas_internet2.csv")
df1.head()
```

```python
df2 = pd.read_csv("datasets_merge/Empresas_tamano2.csv")
df2.head()
```

```python
df3 = pd.read_csv("datasets_merge/Servicios Basicos 2.csv")
df3.head()
```

```python
df4 = pd.read_csv("datasets_merge/datos2.csv")
df4.head()
```

```python
df5 = df0.merge(df1, left_on='Unnamed: 0', right_on='Unnamed: 0')
df5 = df5.merge(df2, left_on='Unnamed: 0', right_on='Unnamed: 0')
df5 = df5.merge(df3, left_on='Unnamed: 0', right_on='Unnamed: 0')
df5 = df5.merge(df4, left_on='Unnamed: 0', right_on='Unnamed: 0')
df5.head()
```

```python
df5 = df5.set_index("Unnamed: 0")
```

```python
df5.to_csv("final.csv")
```
