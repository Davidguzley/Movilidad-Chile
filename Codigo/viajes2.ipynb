```python
import pandas as pd
import numpy as np

df = pd.read_csv('matriz_viajes_v2.csv')
```

```python
df["timestamp"] = pd.to_datetime(df["timestamp"])
```

```python
df = df.fillna(0)
```

```python
df = df.drop(['Unnamed: 0'], axis = 1)
```

```python
def tiemp(t1, t2):
    return (t1 - t2) / pd.Timedelta(hours = 1)
```

```python
df
```

```python
aux = {'PHONE_ID': [],
       'Coordenada': [],
       'Hora_entrada': [],
       'Hora_salida': [],
       'Comuna': [],
       'Distancia': [],
       'Tiempo_en_comuna': []
}

primer_registro = 1

for i in range(len(df)):
    if i == len(df) - 1:
        tiempo_fin = df['timestamp'][i]
        aux['Hora_salida'].append(tiempo_fin)
        aux['Tiempo_en_comuna'].append(tiemp(tiempo_fin, aux['Hora_entrada'][len(aux['Hora_entrada']) - 1]))
        
    else:
        if df['PHONE_ID'][i] == df['PHONE_ID'][i + 1]:
            if df['coordinates'][i] == df['coordinates'][i + 1]:
                if primer_registro == 1:
                    primer_registro = 0
                    aux['PHONE_ID'].append(df['PHONE_ID'][i])
                    aux['Coordenada'].append(df['coordinates'][i])
                    aux['Hora_entrada'].append(df['timestamp'][i])
                    aux['Comuna'].append(df['comuna'][i])
                    aux['Distancia'].append(0)
                
            else:
                if primer_registro == 1:
                    primer_registro = 0
                    aux['PHONE_ID'].append(df['PHONE_ID'][i])
                    aux['Coordenada'].append(df['coordinates'][i])
                    aux['Hora_entrada'].append(df['timestamp'][i])
                    aux['Comuna'].append(df['comuna'][i])
                    aux['Distancia'].append(0)
                    tiempo_fin = df['timestamp'][i]
                    aux['Hora_salida'].append(tiempo_fin)
                    aux['Tiempo_en_comuna'].append(0)
                    
                    aux['PHONE_ID'].append(df['PHONE_ID'][i + 1])
                    aux['Coordenada'].append(df['coordinates'][i + 1])
                    aux['Hora_entrada'].append(df['timestamp'][i + 1])
                    aux['Comuna'].append(df['comuna'][i + 1])
                    aux['Distancia'].append(df['dis_trans'][i + 1])
                    
                else:
                    tiempo_fin = df['timestamp'][i]
                    aux['Hora_salida'].append(tiempo_fin)
                    aux['Tiempo_en_comuna'].append(tiemp(tiempo_fin, aux['Hora_entrada'][len(aux['Hora_entrada']) - 1]))


                    aux['PHONE_ID'].append(df['PHONE_ID'][i + 1])
                    aux['Coordenada'].append(df['coordinates'][i + 1])
                    aux['Hora_entrada'].append(df['timestamp'][i + 1])
                    aux['Comuna'].append(df['comuna'][i + 1])
                    aux['Distancia'].append(df['dis_trans'][i + 1])
                
        else:
            if df['PHONE_ID'][i - 1] != df['PHONE_ID'][i]:
                aux['PHONE_ID'].append(df['PHONE_ID'][i])
                aux['Coordenada'].append(df['coordinates'][i])
                aux['Hora_entrada'].append(df['timestamp'][i])
                aux['Comuna'].append(df['comuna'][i])
                aux['Distancia'].append(0)
                tiempo_fin = df['timestamp'][i]
                aux['Hora_salida'].append(tiempo_fin)
                aux['Tiempo_en_comuna'].append(0)
                
            else:
                tiempo_fin = df['timestamp'][i]
                aux['Hora_salida'].append(tiempo_fin)
                aux['Tiempo_en_comuna'].append(tiemp(tiempo_fin, aux['Hora_entrada'][len(aux['Hora_entrada']) - 1]))
            primer_registro = 1
```
